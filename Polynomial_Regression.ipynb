```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import the dataset

data = pd.read_csv("https://raw.githubusercontent.com/SairamTadepalli19/Day6_ML/main/Position_Salaries.csv")

x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

```python
#training the linear regression 

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(x, y)
```

```python
#Implement polynomial regression on whole dataset

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)

x_poly = poly_reg.fit_transform(x)

lr2 = LinearRegression()

lr2.fit(x_poly, y)
```

```python
#Visualizing Linear Regression

plt.scatter(x,y, color = 'red')
plt.plot(x, lr.predict(x), color = 'blue')
plt.title("Level Vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")

plt.show()
```

```python
#Visualizing Linear Regression

plt.scatter(x,y, color = 'red')
plt.plot(x, lr2.predict(poly_reg.transform(x)), color = 'blue')
plt.title("Level Vs Salary")
plt.xlabel("Level")
plt.ylabel("Salary")

plt.show()
```

```python
#let's predict a new value using linear regression

lr.predict([[6.5]])
```

```python
lr.predict([[5]])
```

```python
#Let's predict a new value using polynomial regression

lr2.predict(poly_reg.fit_transform([[6.5]]))
```
